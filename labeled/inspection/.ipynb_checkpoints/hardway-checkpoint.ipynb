```python
import numpy as np
import pandas as pd

# Assuming you have your DataFrame 'DF' loaded with participant and sip IDs
# Prepare lists to hold padded sequences and target volumes
padded_sequences = []
target_volumes = []

# Iterate through unique participant IDs
for participant_id in DF['Participant_ID'].unique():
    participant_data = DF[DF['Participant_ID'] == participant_id]
    
    # Find the maximum time steps for each specific participant
    max_time_steps = participant_data['sip_id'].apply(lambda sip_id: len(participant_data[participant_data['sip_id'] == sip_id])).max()
    print(f"Participant ID: {participant_id} | Maximum Time Steps: {max_time_steps}")

    # Iterate through each sip ID for the participant
    for sip_id in participant_data['sip_id'].unique():
        sip_data = participant_data[participant_data['sip_id'] == sip_id]
        time_steps = len(sip_data)

        # Extract zoning data (assuming you have 8 Zones)
        zones = sip_data[[f'Zone_{i}' for i in range(8)]].values
        
        # Extract TOF data; replace this with your actual data extraction logic
        toft_data = np.random.rand(time_steps, 64)  # Placeholder for TOF data
        print(f"Participant ID: {participant_id} | Sip ID: {sip_id} | Time Steps: {time_steps} | Zones Shape: {zones.shape} | TOF Shape: {toft_data.shape}")

        # Padding for zones
        if time_steps < max_time_steps:
            zone_padding = np.zeros((max_time_steps - time_steps, zones.shape[1]))  # Padding for zones
            padded_zones = np.vstack([zones, zone_padding])
        else:
            padded_zones = zones[-max_time_steps:]  # Take the last max_time_steps values
        
        # Padding for TOF data
        if time_steps < max_time_steps:
            tof_padding = np.zeros((max_time_steps - time_steps, 64))  # 64 TOF features
            padded_tofs = np.vstack([toft_data, tof_padding])
        else:
            padded_tofs = toft_data[-max_time_steps:]  # Take the last max_time_steps values

        print(f"Padded Zones Shape: {padded_zones.shape} | Padded TOF Shape: {padded_tofs.shape}")

        # Combine all features into one array; ensure you flatten the zones and TOF data
        combined_data = np.hstack([
            padded_zones.flatten(),  # Flatten zones
            padded_tofs.flatten(),   # Flatten TOF data
            [sip_data['Container_Weight'].values[0], 
             sip_data['drink'].values[0], 
             sip_data['temp'].values[0], 
             participant_id]          # Include additional features
        ])
        
        padded_sequences.append(combined_data)  # Append to the final list
        
        # Calculate mean volume for the target
        target_volumes.append(np.mean(sip_data['dV'].values))

# Convert padded_sequences to numpy array and target volumes
padded_sequences = np.array(padded_sequences)
target_volumes = np.array(target_volumes)

# Print final shapes before preparing the model
print(f"Final Padded Sequences Shape: {padded_sequences.shape}")
print(f"Final Target Volumes Shape: {target_volumes.shape}")

# Prepare the data for modeling
# Reshape input data for LSTM, assuming combined data is flat
# Ensure proper number of features based on handling
features_per_timestep = (8 + 64)  # Zones (8) + TOF data (64)

# Reshape the padded_sequences to (num_samples, max_time_steps, features_per_timestep)
X = padded_sequences.reshape(-1, max_time_steps, features_per_timestep + 4)  # Including feature counts
y = target_volumes

# Model preparation
import keras
from keras.models import Sequential
from keras.layers import Conv1D, LSTM, Dense, Flatten, Dropout

# Define the model
model = Sequential()
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(max_time_steps, features_per_timestep + 4)))
model.add(LSTM(50, return_sequences=False))  # Adjusting for sequence processing
model.add(Dropout(0.5))
model.add(Dense(1, activation='linear'))  # Adjust according to your output shape if needed

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.summary()  # View the model structure

# Train the model
model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)
```

```python
import numpy as np
import pandas as pd

# Initialize a placeholder DataFrame 'DF' 
# Replace this with your actual DataFrame that contains participant data

# Sample code initializing DataFrame, ensure you replace it with your data
# Add your actual data frame loading logic here

# Initialize placeholder lists
padded_sequences = []
target_volumes = []

# Iterate through unique participant IDs
for participant_id in DF['Participant_ID'].unique():
    participant_data = DF[DF['Participant_ID'] == participant_id]
    
    # Determine maximum time steps for this participant
    max_time_steps = participant_data['sip_id'].apply(
        lambda sip_id: len(participant_data[participant_data['sip_id'] == sip_id])
    ).max()
    
    print(f"Participant ID: {participant_id} | Maximum Time Steps: {max_time_steps}")

    # Iterate through each sip ID for the participant
    for sip_id in participant_data['sip_id'].unique():
        sip_data = participant_data[participant_data['sip_id'] == sip_id]
        time_steps = len(sip_data)

        # Extract zones and TOF data
        zones = sip_data[[f'Zone_{i}' for i in range(8)]].values  # Zones shape (time_steps, 8)
        toft_data = np.random.rand(time_steps, 64)  # Placeholder for TOF; use actual data here

        print(f"Participant ID: {participant_id} | Sip ID: {sip_id} | Time Steps: {time_steps}")

        # Handle padding for zones
        if time_steps < max_time_steps:
            zone_padding = np.zeros((max_time_steps - time_steps, zones.shape[1]))  # Zone padding shape
            padded_zones = np.vstack((zones, zone_padding))  # Shape: (max_time_steps, 8)
        else:
            padded_zones = zones[-max_time_steps:]  # Shape: (max_time_steps, 8)

        # Handle padding for TOF data
        if time_steps < max_time_steps:
            tof_padding = np.zeros((max_time_steps - time_steps, 64))  # TOF padding shape
            padded_tofs = np.vstack((toft_data, tof_padding))  # Shape: (max_time_steps, 64)
        else:
            padded_tofs = toft_data[-max_time_steps:]  # Shape: (max_time_steps, 64)

        print(f"Padded Zones Shape: {padded_zones.shape} | Padded TOF Shape: {padded_tofs.shape}")

        # Combine features into a single flat array
        gender = sip_data['Gender'].values[0]  # Get gender value
        container_weight = sip_data['Container_Weight'].values[0]  # Get container weight value

        # Combine into a single flat array
        combined_data = np.hstack([
            padded_zones.flatten(),  # Flattened shape: (max_time_steps * 8,)
            padded_tofs.flatten(),    # Flattened shape: (max_time_steps * 64,)
            [container_weight, gender]  # Additional features
        ])

        # Determine expected length
        expected_length = (max_time_steps * (8 + 64)) + 2  # 8 from zones, 64 from TOF + 2 features (weight and gender)
        
        # Append combined data if the shape matches
        if combined_data.shape[0] == expected_length:
            padded_sequences.append(combined_data)  # Append only if valid
        else:
            print(f"Warning: unexpected shape for combined data: {combined_data.shape} for Participant {participant_id}, Sip {sip_id}.")

        # Append mean volume for target array
        target_volumes.append(np.mean(sip_data['dV'].values))

# Attempt to convert padded_sequences to a NumPy array
try:
    if padded_sequences:
        padded_sequences = np.array(padded_sequences)  # Convert from list to NumPy array
        target_volumes = np.array(target_volumes)
        print("Padded Sequences Shape:", padded_sequences.shape)
    else:
        print("No sequences were generated.")

    # Check reshaping requirements
    if padded_sequences.ndim == 2:
        X = padded_sequences.reshape(-1, max_time_steps, padded_sequences.shape[1])  # For CNN/LSTM input
        y = target_volumes

        # Define your CNN model or LSTM model here
        import keras
        from keras.models import Sequential
        from keras.layers import LSTM, Dense, Dropout

        model = Sequential()
        model.add(LSTM(50, input_shape=(max_time_steps, padded_sequences.shape[1]), return_sequences=False))  # LSTM layer
        model.add(Dropout(0.5))  # Dropout layer to prevent overfitting
        model.add(Dense(1, activation='linear'))  # Output layer

        model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
        model.summary()

        # Train the model
        model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)
    else:
        print("Padded sequences do not have the correct shape to reshape for input data.")

except ValueError as e:
    print(f"Error converting padded_sequences to array: {e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error
import math

# Sample DataFrame loading; replace this with your actual DataFrame
# DF = pd.read_csv('your_data_file.csv')

# Padded sequences and targets
padded_sequences = []
target_volumes = []

# Loop through each participant
for participant_id in DF['Participant_ID'].unique():
    participant_data = DF[DF['Participant_ID'] == participant_id]
    
    # Determine maximum time steps for the participant
    max_time_steps = participant_data['sip_id'].apply(
        lambda sip_id: len(participant_data[participant_data['sip_id'] == sip_id])
    ).max()

    # Iterate through each sip for the participant
    for sip_id in participant_data['sip_id'].unique():
        sip_data = participant_data[participant_data['sip_id'] == sip_id]
        time_steps = len(sip_data)

        # Extract zones and TOF data
        zones = sip_data[[f'Zone_{i}' for i in range(8)]].values  # shape: (time_steps, 8)
        toft_data = np.random.rand(time_steps, 64)  # Placeholder for TOF data; replace with actual TOF data

        # Determine the shapes before padding
        original_zones_shape = zones.shape
        original_tofs_shape = toft_data.shape

        # Handle padding for zones
        if time_steps < max_time_steps:
            zone_padding = np.zeros((max_time_steps - time_steps, zones.shape[1]))  # Padding shape for zones
            padded_zones = np.vstack((zones, zone_padding))  # (max_time_steps, 8)
        else:
            padded_zones = zones[-max_time_steps:]  # Use the latest max_time_steps

        # Handle padding for TOF data
        if time_steps < max_time_steps:
            tf_padding = np.zeros((max_time_steps - time_steps, 64))  # Padding for TOF
            padded_tofs = np.vstack((toft_data, tf_padding))  # (max_time_steps, 64)
        else:
            padded_tofs = toft_data[-max_time_steps:]  # Use the latest max_time_steps

        # Combine features into a single array
        gender = sip_data['Gender'].values[0]  # Extract gender
        container_weight = sip_data['Container_Weight'].values[0]  # Extract weight

        combined_data = np.hstack([
            padded_zones.flatten(),  # Shape: (max_time_steps * 8,)
            padded_tofs.flatten(),   # Shape: (max_time_steps * 64,)
            [container_weight, gender]  # Additional features
        ])

        # Calculate expected length
        expected_length = (max_time_steps * (8 + 64)) + 2  # 8 zones + 64 TOF + 2 features (weight & gender)

        # Append combined data if its shape is as expected
        if combined_data.shape[0] == expected_length:
            padded_sequences.append(combined_data)
            target_volumes.append(np.mean(sip_data['dV'].values))  # Capture dV mean for this sip
        else:
            print(f"Warning: Unexpected shape {combined_data.shape} for Participant {participant_id}, Sip {sip_id}.")

# Ensure padded_sequences is converted only if it has the right shape
if padded_sequences:
    try:
        # Convert lists to NumPy array
        padded_sequences = np.array(padded_sequences)
        target_volumes = np.array(target_volumes)
        print("Padded Sequences Shape:", padded_sequences.shape)

        # Reshape input for the neural network
        if padded_sequences.ndim == 2 and padded_sequences.shape[0] > 0:
            X = padded_sequences.reshape(-1, max_time_steps, padded_sequences.shape[1])
            y = target_volumes

            # Define the LSTM model
            model = Sequential()
            model.add(LSTM(50, input_shape=(max_time_steps, padded_sequences.shape[1]), return_sequences=False))
            model.add(Dropout(0.5))
            model.add(Dense(1, activation='linear'))

            # Compile the model
            model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
            model.summary()

            # Fit the model
            history = model.fit(X, y, epochs=50, batch_size=32, validation_split=0.2)

            # Make predictions
            predictions = model.predict(X)

            # Evaluate the model with MSE and RMSE
            mse = mean_squared_error(y, predictions)
            rmse = math.sqrt(mse)

            # Display the results
            print(f"Mean Squared Error (MSE): {mse:.4f}")
            print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")

            # Plotting Actual vs Predicted values
            plt.figure(figsize=(10, 6))
            plt.scatter(y, predictions, color='blue', alpha=0.5)
            plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Ideal predictor line
            plt.title('Actual vs Predicted')
            plt.xlabel('Actual Volume (dV)')
            plt.ylabel('Predicted Volume (dV)')
            plt.xlim([y.min(), y.max()])
            plt.ylim([y.min(), y.max()])
            plt.grid()
            plt.show()

        else:
            print("Padded sequences do not have the correct shape to reshape for input data.")
    except ValueError as e:
        print(f"Error converting padded_sequences to array: {e}")
else:
    print("No sequences were generated.")
```
