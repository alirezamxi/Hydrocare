```python
import os

# Input MJPEG file
input_file = 'image.jpg'
output_dir = 'frames_extracted'

# JPEG start and end markers
SOI = b'\xff\xd8'  # Start of Image
EOI = b'\xff\xd9'  # End of Image

# Make output directory
os.makedirs(output_dir, exist_ok=True)

with open(input_file, 'rb') as f:
    data = f.read()

frame_count = 0
index = 0

while True:
    start = data.find(SOI, index)
    if start == -1:
        break

    end = data.find(EOI, start)
    if end == -1:
        break

    # Extract JPEG bytes
    jpg_data = data[start:end + 2]

    # Save to file
    frame_filename = os.path.join(output_dir, f'frame_{frame_count:05d}.jpg')
    with open(frame_filename, 'wb') as img_file:
        img_file.write(jpg_data)

    print(f"✅ Extracted frame {frame_count} -> {frame_filename}")
    frame_count += 1
    index = end + 2

print(f"🎉 Done! Extracted {frame_count} frames into '{output_dir}/'")
```
